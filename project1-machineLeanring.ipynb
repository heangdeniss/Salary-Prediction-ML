```python

import plotly.express as px

fig = px.scatter(
    df,
    x='Years of Experience',
    y='Salary',
    title='Years of Experience vs Salary',
    labels={'YearsExperience': 'Years of Experience', 'Salary': 'Salary ($)'},
    hover_data=['Years of Experience', 'Salary']
)

fig.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python

# Split the dataset into training and testing sets
feature = df[['Years of Experience', 'Education Level']]
target = df['Salary']
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)

# Encode the categorical feature "Education Level" from String to float in order to make prediction since we cant predict using String
encoder = OneHotEncoder(sparse_output=False)
encoded_education = encoder.fit_transform(X_train[['Education Level']])

encoded_education = pd.DataFrame(
    encoded_education, 
    columns=encoder.get_feature_names_out(['Education Level']),
    index=X_train.index
)

X_train_encoded = pd.concat([X_train.drop('Education Level', axis=1), encoded_education], axis=1)

# Encode education Test
encoded_education_test = encoder.transform(X_test[['Education Level']])
encoded_education_test = pd.DataFrame(
    encoded_education_test,
    columns=encoder.get_feature_names_out(['Education Level']),
    index=X_test.index
)
X_test_encoded = pd.concat([X_test.drop('Education Level', axis=1), encoded_education_test], axis=1)

# Train the Linear Regression model with the encoded features
model = LinearRegression()
model.fit(X_train_encoded, y_train)
```

```python
y_pred = model.predict(X_test_encoded)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R-squared: {r2:.2f}")
```

### MSE: Average squared errors.
### RMSE: Average prediction error in the same units as salary.
### R²: Model performance indicator (closer to 1 means better fit).

```python
import plotly.express as px
import pandas as pd

# Prepare a clear results DataFrame
results_df = pd.DataFrame({
    'YearsExperience': X_test_encoded['Years of Experience'],
    'Actual Salary': y_test,
    'Predicted Salary': y_pred
})

fig = px.scatter(results_df, x='YearsExperience', y='Actual Salary',
                 labels={'YearsExperience':'Years of Experience', 'Actual Salary':'Salary ($)'},
                 title='Actual vs Predicted Salary (Interactive Scatter)',
                 hover_data=['Actual Salary', 'Predicted Salary'])

# predict salary point
fig.add_scatter(x=results_df['YearsExperience'], y=results_df['Predicted Salary'],
                mode='markers',
                marker=dict(color='red'),
                name='Predicted Salary')

fig.update_layout(legend_title_text='Salary Type',
                  legend=dict(y=0.99, x=0.01),
                  hovermode='closest')

fig.show()
```
